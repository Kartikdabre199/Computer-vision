## Sharpen the image 

```python
train= cv2.imread(r"G:\cnn\some images\train7.PNG").astype(np.float32)/255
train= cv2.cvtColor(train, cv2.COLOR_BGR2RGB)
```

```python
train7 = plt.imshow(train)
```

```python
sharp_blur_factor = 3
# darken the image 
```

```python
train1 = np.power(train, sharp_blur_factor)
```

```python
plt.imshow(train1)
```

The image is much more sharpen then the original

## Bluring the image

```python
sharp_blur_factor = 1/2

train1 = np.power(train, sharp_blur_factor)

plt.imshow(train1)
```
